```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def display_img(img,cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax  = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
reeses = cv2.imread('DATA/reeses_puffs.png',0)
```

```python
display_img(reeses)
```

```python
cereals = cv2.imread('DATA/many_cereals.jpg',0)
```

```python
display_img(cereals)
```

```python
orb = cv2.ORB_create()
```

```python
kp1,ds1 = orb.detectAndCompute(reeses,None)
kp2,ds2 = orb.detectAndCompute(cereals,None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck = True)
```

```python
matches = bf.match(ds1,ds2)
single_match = matches[0]
single_match.distance
```

```python
matches = sorted(matches,key = lambda x:x.distance)
```

```python
reeses_matches = cv2.drawMatches(reeses,kp1,cereals,kp2,matches[:25],None,flags=2)
```

```python
display_img(reeses_matches)
```

```python
sift = cv2.xfeatures2d.SIFT_create()
```

```python
kp1,ds1 = sift.detectAndCompute(reeses,None)
kp2,ds2 = sift.detectAndCompute(cereals,None)
```

```python
bf = cv2.BFMatcher()
```

```python
matches = bf.knnMatch(ds1,ds2,k=2)
```

```python
matches
```

```python
good = []

# Ratio Test

for match1,match2 in matches:
    if match1.distance < 0.75*match2.distance:
        good.append([match1])
```

```python
len(good)
```

```python
len(matches)
```

```python
sift_matches = cv2.drawMatchesKnn(reeses,kp1,cereals,kp2,good,None,flags=2)
```

```python
display_img(sift_matches)
```
