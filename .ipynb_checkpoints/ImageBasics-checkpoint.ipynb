```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('DATA/00-puppy.jpg')
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
while True:
    cv2.imshow('Puppy',img)
    if cv2.waitKey(1) & 0xFF == 27:
        break
        
cv2.destroyAllWindows()
```

```python
img1 = np.zeros((512,512,3),dtype = np.int32)
```

```python
plt.imshow(img1)
```

```python
cv2.rectangle(img1,pt1 = (384,10),pt2 = (500,150),color = (0,255,0),thickness =10)
plt.imshow(img1)
```

```python
cv2.circle(img1,center = (100,100),radius = 50,color = (255,0,0),thickness = 8)
plt.imshow(img1)
```

```python
cv2.circle(img1,center = (100,100),radius = 50,color = (255,0,0),thickness = 8)
plt.imshow(img1)
```

```python
cv2.line(img1,pt1=(0,0),pt2 =(512,512),color = (102,255,255),thickness = 5)
plt.imshow(img1)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img1,text = "OpenCV",org=(10,500),fontFace=font,fontScale = 4,color = (255,255,255),thickness =3,lineType = cv2.LINE_AA)
plt.imshow(img1)
```

```python
img2 = np.zeros((512,512,3),dtype = np.int32)
plt.imshow(img2)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype = np.int32)
```

```python
vertices.shape
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
pts.shape
```

```python
cv2.polylines(img2,[pts],isClosed=True,color = (0,0,255),thickness = 10)
plt.imshow(img2)
```
