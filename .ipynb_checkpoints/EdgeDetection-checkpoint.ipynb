```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('DATA/sammy_face.jpg')
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(img,threshold1 = 127,threshold2 = 127)
```

```python
plt.imshow(edges)
```

```python
edges = cv2.Canny(img,0,255)
plt.imshow(edges)
```

```python
med_value = np.median(img)
```

```python
med_value
```

```python
# Lower threshold to either 0 or 70% of the median value whichever is greater.
lower = int(max(0,0.7*med_value))
# Upper threshold to either 130% of the median or the max 255, whichever is smaller
upper = int(min(255,1.3*med_value))
```

```python
edges = cv2.Canny(img,lower,upper+100)
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(img, ksize = (5,5))
```

```python
edges = cv2.Canny(blurred_img,lower,upper+100)
plt.imshow(edges)
```
