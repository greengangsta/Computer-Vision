```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
dark_horse = cv2.imread('DATA/horse.jpg')  # Original BGR channel coding
show_horse = cv2.cvtColor(dark_horse,cv2.COLOR_BGR2RGB) # RGB coding for matplotlib

rainbow = cv2.imread('DATA/rainbow.jpg')
show_rainbow = cv2.cvtColor(rainbow,cv2.COLOR_BGR2RGB)

blue_bricks = cv2.imread('DATA/bricks.jpg')
show_bricks = cv2.cvtColor(blue_bricks,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(show_horse)
```

```python
plt.imshow(show_rainbow)
```

```python
plt.imshow(show_bricks)
```

```python
hist_values = cv2.calcHist([blue_bricks],channels=[0],mask = None,histSize=[256],ranges=[0,256])
```

```python
hist_values.shape
```

```python
plt.plot(hist_values)
```

```python
hist_values_horse = cv2.calcHist([dark_horse],channels=[0],mask = None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist_values_horse)
```

```python
img  = blue_bricks
color = ('b', 'g', 'r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],mask=None,histSize=[256],ranges=[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])

plt.title('Histogram for blue bricks')
```

```python
img  = dark_horse
color = ('b', 'g', 'r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],mask=None,histSize=[256],ranges=[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,50])
    plt.ylim([0,500000])

plt.title('Histogram for dark horse')
```

```python
img  = rainbow
color = ('b', 'g', 'r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],mask=None,histSize=[256],ranges=[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])

plt.title('Histogram for rainbow')
```

```python
rainbow = cv2.imread('DATA/rainbow.jpg')
show_rainbow = cv2.cvtColor(rainbow,cv2.COLOR_BGR2RGB)
```

```python
img = rainbow
```

```python
img.shape
```

```python
mask = np.zeros(img.shape[:2],dtype = np.uint8)
```

```python
plt.imshow(mask,cmap='gray')
```

```python
mask[300:400,100:400] = 255
```

```python
plt.imshow(mask,cmap='gray')
```

```python
plt.imshow(show_rainbow)
```

```python
masked_img = cv2.bitwise_and(img,img,mask = mask)
```

```python
masked_img_show  = cv2.bitwise_and(show_rainbow,show_rainbow,mask=mask)
```

```python
plt.imshow(masked_img_show)
```

```python
hist_masked_values_red = cv2.calcHist([rainbow],channels = [2],mask = mask,histSize = [256],ranges = [0,256])
```

```python
hist_values_red = cv2.calcHist([rainbow],channels = [2],mask = None,histSize = [256],ranges = [0,256])
```

```python
plt.plot(hist_masked_values_red)
plt.title('Red histogram for masked rainbow')
```

```python
plt.plot(hist_values_red)
plt.title('Red Histogram for Normal Rainbow')
```

```python
gorilla = cv2.imread('DATA/gorilla.jpg',0)
```

```python
def display_img(img,cmap):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
display_img(gorilla,cmap='gray')
```

```python
gorilla.shape
```

```python
hist_values = cv2.calcHist([gorilla],channels=[0],mask = None,histSize = [256],ranges = [0,256])
```

```python
plt.plot(hist_values)
```

```python
eq_gorilla = cv2.equalizeHist(gorilla)
```

```python
display_img(eq_gorilla,'gray')
```

```python
hist_values_eq = cv2.calcHist([eq_gorilla],channels=[0],mask = None,histSize = [256],ranges = [0,256])
```

```python
plt.plot(hist_values_eq)
```

```python
color_gorilla = cv2.imread('DATA/gorilla.jpg')
```

```python
img = cv2.cvtColor(color_gorilla,cv2.COLOR_BGR2RGB)
```

```python
display_img(img,None)
```

```python
hsv_gorilla = cv2.cvtColor(color_gorilla,cv2.COLOR_BGR2HSV)
```
